```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
df = df.rename(columns={'per capita income (US$)': 'new_label'})
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capital Income')
plt.scatter(df.year,df.new_label,color='red',marker='+')
```

```python
new_df = df.drop('new_label',axis='columns')
new_df
```

```python
new_price = df.drop('year',axis='columns')
new_price
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,new_price)
```

```python
reg.predict([[2018]])
```

Y = m * X + b (m is coefficient and b is intercept) **bold text**

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capital Income')
plt.scatter(df.year,df.new_label,color='red',marker='+')
plt.plot(df.year, reg.predict(df[['year']]),color='black')
```
